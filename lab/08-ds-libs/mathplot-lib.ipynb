# Basics

## Einführung

Die `matplotlib` ist das Python Modul zur grafischen Darstellung von Daten.
* Es ist ähnlich zu den Grafischen Funktionen von Matlab
* `matplotlib.pyplot` enthält die Funktionen um Graphen zu erstellen.

Es wird oft als `plt` importiert:
```
import matplotlib.pyplot as plt
```

Bei der Darstellung von matplotlib innerhalb eines Yupiter Noteobooks,
muss die folgende Linie angegeben werden:
```
%matplotlib inline
```

**Beispiel**
* Die `plt.figure` Funktion generiert eine neue Figur
* Mit der `plt.plot` Funktion kann man Daten plotten.
* `plt.show` zeigt den plot auf dem Bildschirm
* `plt.savefig` speichert die Figur
* `plt.close` schliesst die Figur

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Daten vorbereiten
x = np.arange(-5,5.01,0.5)
y1 = 1*x + 1.5 +np.random.normal(0, 1, len(x))
y2 = 2*x +np.random.normal(0, 1, len(x))

print(x)
print(y1)

# Grafik plotten
plt.figure()
plt.plot(x, y1)
plt.show()
# plt.savefig("output/plot.png")
plt.close()
````

## Formatierung

**Linienformate (plt.plot)**

Neben y und y können optionale Argumente für die Darstellung der Linie und Punkte definiert werden:
- marker="x", 
- color="g"
- linestyle="--",
- markersize=10
- linewidth=4
- label="y" (für Darstellung in der Legende mit plt.legend)


**Titel und Achsen (plt.title, plt.xlabel, plt.ylabel)**

````python
plt.figure()
plt.plot(x, y1)
plt.xlabel("Punkte ABC", fontsize=24)
plt.ylabel("y-label", fontsize=18)
plt.title("My plot title", fontsize=40)
plt.show()
plt.close()
````

**Legende (plt.legend)**

Darstellung der Legende mit Label von plt.plot Befehl:

````python
plt.figure()
plt.plot(x, y1, "ro", label="y1")
plt.plot(x, y2, "g--", label="y2")
plt.legend(loc="best")
plt.show()
plt.close()
````

**Latex**

In allen Text Felder kann man auch die Latex syntax brauchen um Mathematische symbolen zu representieren. Einfach das Latex zwischen $ zeichen:

````python
plt.figure()
plt.plot(x, y1)
plt.xlabel("distance in [$\AA$]")
plt.ylabel("$a*x+b$")
plt.title("My function $\sqrt{f_a(x)}$")
plt.show()
plt.close()
````

**Figur Grösse (plt.figure)**
Die grösse der Figur wird in plt.figure() mit figsize angegeben:
```
plt.figure(figsize = (8, 4))     # [inches]
```

**Achsen Limiten (plt.xlim, plt.ylim, plt.xscale)**
- plt.xlim und plt.ylim um der range von den Achsen zu kontrollieren
- plt.xscale("log") für ein logaritmischer Achsen
```
plt.xlim((1, 5))           
plt.ylim((-10, 10))
plt.xscale("log")
```

## Linien, Texte und Annotationen

**Linien (plt.vlines, plt.hlines)**

Mit `plt.vlines` und `plt.hlines` können zusätzliche vertikale und horizontale Linien hinzugefügt werden. Dabei werden folgende Argumente definiert:
- Arg 1: Position oder Liste von Positionen
- Arg 2: Start und End Position
- Arg n: Weiter Argumente wie linestyle, color, etc.

````python
norm_dist1 = np.random.normal(0, 1, 1000)
norm_dist2 = np.random.normal(1, 0.5, 1000)
plt.hist([norm_dist1,norm_dist2], bins=20, density=True)
plt.hlines(0.5, plt.xlim()[0], plt.xlim()[1], linestyle=":", color="r")
plt.vlines([0,1.0],*plt.ylim(),linestyle = "--")
plt.show()
plt.close()
````

**Text und Annotationen (plt.text, plt.annotate)**

Hinzufügen von Text und Annotationen:
- `plt.text(xpos, ypos, text)` 
  - um text hinzufügen an der Position `(xpos, ypos)`
- `plt.annotate(text, xy=(xpos, ypos), xytext=(text_x, text_y))` 
  - um Text hinzufügen an der Position `(text_x, text_y)` 
  - mit einem Pfeil der auf `(xpos, ypos)` zeigt.


````python
plt.figure()
plt.plot(x, y1,"x")
plt.text(0,0, "text at position 0,0", fontsize=16)
plt.annotate("point to 1,1",xy=(1, 1), xytext=(3, 2),arrowprops=dict(facecolor='black'))
plt.show()
plt.close()
````

## Format Strings, Markers, Line Styles, Colors

### Format Strings

A format string (fmt) consists of a part for color, marker and line::
> fmt = '[marker][line][color]'

- Each of them is optional. If not provided, the value from the style cycle is used. 
- Exception: If ``line`` is given, but no ``marker``, the data will be a line without markers.
- Other combinations such as ``[color][marker][line]`` are alsocsupported, but note that their parsing may be ambiguous.

**Markers**

character  | marker                   | character  | marker
---------- | ------------------------ | ---------- | -------------------------------------
'.'        | point                    | 's'        | square 
','        | pixel                    | 'p'        | pentagon
'o'        | circle                   | '*'        | star 
'v'        | triangle_down            | 'h'        | hexagon1 
'^'        | triangle_up              | 'H'        | hexagon2 
'<'        | triangle_left            | '+'        | plus 
'>'        | triangle_right           | 'x'        | x
'1'        | tri_down                 | 'D'        | diamond 
'2'        | tri_up                   | 'd'        | thin_diamond
'3'        | tri_left                 | '|'        | vline
'4'        | tri_right                | '_'        | hline 
    
**Line Styles**
character  | line style   
---------- | ------------------------ 
'-'        | solid 
'--'       | dashed
'-.'       | dash-dot
':'        | dotted 

**Colors**
character  | color
---------- | ------------------------ 
'b'        | blue
'g'        | green
'r'        | red
'c'        | cyan
'm'        | magenta
'y'        | yellow
'k'        | black
'w'        | white

If the color is the only part of the format string, you can additionally use any  `matplotlib.colors` spec, e.g. full names (like 'green') or hex strings (like '#008000').

**Example format strings**
```
'b'    # blue markers with default shape
'or'   # red circles
'-g'   # green solid line
'--'   # dashed line with default color
'^k:'  # black triangle_up markers connected by a dotted line
```

```

mpl.rcParams['font.size'] = 18
mpl.rcParams['legend.fontsize'] = 16

mpl.rcParams['xtick.major.size'] = 12
plt.hist([norm_dist1,norm_dist2], bins=20, density=True)
```

# Errorbar 

Errorbars werden mit plt.errorbar erstellt:

````python
y1_err = np.random.normal(0, 1, len(x))
x_err = np.random.normal(0, 1, len(x))
plt.figure()
plt.errorbar(x, y1, yerr=y1_err, marker = "o", capsize = 5)
plt.errorbar(x, y2, xerr=x_err, marker = "x", capsize = 6)
plt.show()
plt.close()
````

# Histogramm 

Parameter:
- nbins kontrolliert wieviele bins man brauchen will.
- density kontrolliert ob man die Number von Daten in jedem bin will oder die Propabilität Verteilung
- density=True bedeutet: Summe der BAR Fläche = 1 (Propabilität)

````python
norm_dist = np.random.normal(0, 1, 500)
plt.hist(norm_dist, density=True, bins=40)
plt.show()
plt.close()
````

**Mehrere Datensätze**
Mann kann auch das histogram für mehrere datensätze gleichzeitig rechnen:

````python
norm_dist1 = np.random.normal(0, 1, 1000)
norm_dist2 = np.random.normal(1, 0.5, 1000)
plt.hist([norm_dist1,norm_dist2], bins=20, density=True)
plt.show()
plt.close()
````

# Linien Plot

Die plt.hist funktion gibt die bin Anfang und Ende, die Werte, und die Patches zurück. Damit kann ein normaler Linien Plot erstellt werden:

````python
hist_y, hist_x, p = plt.hist(norm_dist1, bins=20, density=True)
hist_x = 0.5 * (hist_x[1:] + hist_x[:-1])
plt.close()
plt.figure()
plt.plot(hist_x, hist_y, "rx-")
plt.show()
plt.close()
````

# Korrelation Matrix

Sample with boston data:

````python
from sklearn.datasets import load_boston
boston = load_boston()

corr_mat = np.corrcoef(np.concatenate((boston.data, boston.target[:,np.newaxis]),axis=1).T)
labels = np.append(boston.feature_names,"PRICE")

plt.figure()
plt.imshow(corr_mat, vmin=-1, vmax=1)
plt.xticks(range(14),labels, rotation=90)
plt.yticks(range(14),labels)
plt.colorbar()
````

# Subplots
* Eine Figure mit mehreren Plots wird mit `plt.subplots` generiert. Gibt die Figur und eine liste von `axes` zurück
* Dan einfach die elementen von `axes` brauchen um zu plotten
* mit `plt.subplots_adjust` kann man die position von den subplots kontrollieren

Für `axes` Objekten werden mehrere Methoden von `plt` umgenennt: `set_` wird vor dem Namen eingefügt. Zum Beispiel:
* `plt.title` -> `ax.set_title`
* `plt.xlabel` -> `ax.set_xlabel`
* `plt.xlim` -> `ax.set_xlim`


````python
f, axes = plt.subplots(2,2,sharey = "row", figsize = (10,8))
f.suptitle("overall title")
axes[0,0].plot(x,y1)
axes[0,1].errorbar(x, y2, y1_err, capsize = 5)
axes[1,0].hist(norm_dist1, density=True, bins=20)
axes[1,1].hist(norm_dist1, density=True, bins=20, cumulative=True)
for i,axl in enumerate(axes):
    for j,ax in enumerate(axl):
        ax.set_title("subtitle ({},{})".format(i,j))
        ax.set_xlabel("x-label ({},{})".format(i,j))
    axl[0].set_ylabel("y-label")
plt.subplots_adjust(right = 0.98, wspace = 0.02, hspace = 0.25)
plt.show()

````
